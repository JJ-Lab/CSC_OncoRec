## Es un rollo el manejo con fechas, aquí está su lectura y cómo quitar las vacías

```python
nat = pd.NaT
dfhist['FECHA MUESTRA'] = pd.to_datetime(dfhist['FECHA MUESTRA'], infer_datetime_format=True)

# dfhist.loc[dfhist['FECHA MUESTRA']=='NaT','FECHA DX 1L']=''
dfhist['FECHA DX 1L'] = dfhist['FECHA DX 1L'].astype(str)
dfhist.loc[dfhist['FECHA DX 1L']=='NaT','FECHA DX 1L']=''

dfhist['FECHA DX 2L'] = dfhist['FECHA DX 2L'].astype(str)
dfhist.loc[dfhist['FECHA DX 2L']=='NaT','FECHA DX 2L']=''
# dfhist['FECHA DX 1L'] = pd.to_datetime(dfhist['FECHA DX 1L'], infer_datetime_format=True).dt.date

pacientes_recaida1=[]
pacientes_recaida2=[]

for p in pacien_list:
    print('\n\nPaciente:',p)
    for m in dfhist.index.values:
#         print(m, 'FECHA DX 1L:', dfhist.loc[m, 'FECHA DX 1L'], type(dfhist.loc[m, 'FECHA DX 1L']))
    
        if (dfhist.loc[m, 'FECHA DX 1L']!='') and (dfhist.loc[m,'CODIGO PACIENTE']==p) and (p not in pacientes_recaida1) :
            print('encuentro recaída en muestra:',m)
            m_p = list(dfhist['CODIGO PACIENTE']==p)
            m_p_rec = [i for i in range(len(m_p)) if m_p[i]]
            print('lista de otras muestras del paciente',p,':',m_p_rec)
            for i in m_p_rec:                
                m0 = dfhist.index[i]    
                ### .days para pasarlo a type(int)
                dfhist.loc[m0, 'FECHA DX 1L'] = dfhist.loc[m, 'FECHA DX 1L']
                dfhist.loc[m0, 'T_recaida1'] = (datetime.strptime(dfhist.loc[m,'FECHA DX 1L'], '%Y-%m-%d') - dfhist.loc[m0,'FECHA MUESTRA']).days
                print('paciente:',dfhist.loc[m0,'CODIGO PACIENTE'],'muestra:',m0,'t=',dfhist.loc[m0,'FECHA MUESTRA'],'tiempo 1ª recaída:', dfhist.loc[m, 'FECHA DX 1L'], 'días:', dfhist.loc[m0,'T_recaida1'])
            if p not in pacientes_recaida1:
                pacientes_recaida1.append(p)
                
print(pacientes_recaida1)
print('\n\nAhora calculamos tiempos para la 2ª recaída')

for p in pacien_list:
    print('\n\nPaciente:',p)
    for m in dfhist.index.values:
    
        if (dfhist.loc[m, 'FECHA DX 2L']!='') and (dfhist.loc[m,'CODIGO PACIENTE']==p) and (p not in pacientes_recaida2) :
            print('encuentro 2ª recaída en muestra:',m, 'en t=',dfhist.loc[m, 'FECHA DX 2L'])
            m_p = list(dfhist['CODIGO PACIENTE']==p)
            m_p_rec = [i for i in range(len(m_p)) if m_p[i]]
            print('lista de otras muestras del paciente',p,':',m_p_rec)
            for i in m_p_rec:                
                m0 = dfhist.index[i]    
                ### .days para pasarlo a type(int)
                dfhist.loc[m0, 'FECHA DX 2L'] = dfhist.loc[m, 'FECHA DX 2L']
                t1 = datetime.strptime(dfhist.loc[m,'FECHA DX 1L'], '%Y-%m-%d')
                t2 = datetime.strptime(dfhist.loc[m,'FECHA DX 2L'], '%Y-%m-%d')
                t = dfhist.loc[m0,'FECHA MUESTRA']
                print(t1,t2,'t:',t)
                if t>t1:
                    dfhist.loc[m0, 'T_recaida2'] = (t2 - t).days
                    print('paciente:',dfhist.loc[m0,'CODIGO PACIENTE'],'muestra:',m0,'t=',dfhist.loc[m0,'FECHA MUESTRA'],'t_1ª recaída:', t1,'t_2ª recaída:', t2, 'días:', dfhist.loc[m0,'T_recaida2'])
            if p not in pacientes_recaida2:
                pacientes_recaida2.append(p)

print(pacientes_recaida1)
```
